```python
#import sklearn library for train the model
from sklearn.model_selection import train_test_split
```

```python
#from the dataset here we taking inputs as n,p,k,temperature,hunidity,ph,rainfall
x=np.array(df[['N','P','K','temperature','humidity','ph','season','soil texture','rainfall']])
```

```python
#output
y=np.array(df[["label"]])
```

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
x_train,x_test,y_train,y_test=train_test_split(x,y)
model_dt = LogisticRegression()
```

```python
model_dt.fit(x_train,y_train)
```

```python
model_dt.score(x_test,y_test)
```

```python
#KNEIGHBORSCLASSIFIER
from sklearn.neighbors import KNeighborsClassifier
x_train,x_test,y_train,y_test=train_test_split(x,y)
model_dt = KNeighborsClassifier()
```

```python
model_dt.fit(x_train,y_train)
```

```python
model_dt.score(x_test,y_test)
```

```python
#importing trees from sklearn
from sklearn.tree import DecisionTreeClassifier
x_train,x_test,y_train,y_test=train_test_split(x,y)
model_dt = DecisionTreeClassifier()
```

```python
model_dt.fit(x_train,y_train)
```

```python
#finding the accuracy of test
model_dt.score(x_test,y_test)
```

```python
#finding accuracy of train
model_dt.score(x_train,y_train)
```
